```python
CATALOG = "dbx_niw_analytics"  # replace for your environment
SCHEMA = "ztlf"
RAW_VOLUME = "raw_volume"

spark.sql(
    f"""CREATE SCHEMA IF NOT EXISTS {CATALOG}.{SCHEMA}
    COMMENT 'Zero Trust Lakehouse Framework research schema'"""
)
spark.sql(
    f"CREATE VOLUME IF NOT EXISTS {CATALOG}.{SCHEMA}.{RAW_VOLUME}"
)
spark.sql(f"SHOW SCHEMAS IN {CATALOG}").show(truncate=False)
spark.sql(
    f"SHOW VOLUMES IN {CATALOG}.{SCHEMA}"
).show(truncate=False)
```
